# working with web_scrap

````python
from bs4 import BeautifulSoup
import requests

headers = {
    "User-Agent": "Mozilla/5.0"
}

web = requests.get(
    "https://www.ambitionbox.com/list-of-companies?page=1",
    headers=headers
)

soup = BeautifulSoup(web.text, "html.parser")
````

````python
final=pd.DataFrame()
for j in range (1,11):
    url=f"https://www.ambitionbox.com/list-of-companies?page={j}"
    headers = {"User-Agent": "Mozilla/5.0"}
    web = requests.get( url,headers=headers)
    soup = BeautifulSoup(web.text, "html.parser")
    
    name=[]
    ratings=[]
    Type=[]
    review=[]
    salary=[]
    interview=[]
    job=[]
    benifits=[]
    photos=[]
    company=soup.find_all('div', class_='companyCardWrapper__primaryInformation')
    for i in company:
        name.append(i.find('h2').text.strip())
        ratings.append(i.find('div', class_="rating_star_container").text.strip())
        Type.append(i.find('span', class_="companyCardWrapper__interLinking").text.strip())
        
    company=soup.find_all('div', class_="companyCardWrapper__tertiaryInformation")
    for i in company:
        review.append(i.find_all('span',class_="companyCardWrapper__ActionCount")[0].text.strip())
        salary.append(i.find_all('span',class_="companyCardWrapper__ActionCount")[1].text.strip())
        interview.append(i.find_all('span',class_="companyCardWrapper__ActionCount")[2].text.strip())
        job.append(i.find_all('span',class_="companyCardWrapper__ActionCount")[3].text.strip())
        benifits.append(i.find_all('span',class_="companyCardWrapper__ActionCount")[4].text.strip())
        photos.append(i.find_all('span',class_="companyCardWrapper__ActionCount")[5].text.strip())
    d={
    'name':name,
    'ratings':ratings,
    'type|location':Type,
    'review':review,
    'salary':salary,
    'interview':interview,
    'job':job,
    'benfits':benifits,
    'photos':photos
    }
    df=pd.DataFrame(d)

    final=pd.concat([final,df],ignore_index=True)
        
````

````python
final=final.replace('--',np.nan)
cols = ["review","salary","interview","job","benfits"]

final[cols] = final[cols].replace({'L':'e5','k':'e3'}, regex=True).astype(float)

final= final.astype({
    "ratings": "float"
})
final["photos"] = final["photos"].fillna(0).astype(int)
````
